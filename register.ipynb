```python
from tkinter import*
from tkinter import ttk,messagebox
from PIL import ImageTk
import mysql.connector

class Register:
    def __init__(self,root):
        self.root=root
        self.root.title("Registration Window")
        self.root.geometry("1350x700+0+0")
        self.root.config(bg="white")
        
        self.bg=ImageTk.PhotoImage(file="madhu1.jpeg")
        self.lbl_bg=Label(self.root,image=self.bg).place(x=0,y=0,relwidth=1,relheight=1)
        
        self.left=ImageTk.PhotoImage(file="4.jpg")
        left=Label(self.root,image=self.left).place(x=60,y=100,width=320,height=440)
        
        register_frame=Frame(self.root,bd=2,relief=RIDGE,bg="white")
        register_frame.place(x=380,y=100,width=750,height=440)
        
        title=Label(register_frame,text="REGISTER HERE",font=("lucida handwriting",18,"bold"),bg="white").place(x=0,y=10,relwidth=1)
        

        f_name=Label(register_frame,text="First Name",font=("times new roman",13,"bold"),bg="white").place(x=30,y=70)
        self.txt_fname=Entry(register_frame,font=("times new roman",13),bg="#ECECEC") 
        self.txt_fname.place(x=30,y=90,width=250)
        
        
        l_name=Label(register_frame,text="Last Name",font=("times new roman",13,"bold"),bg="white").place(x=350,y=70)
        self.txt_lname=Entry(register_frame,font=("times new roman",13),bg="#ECECEC")
        self.txt_lname.place(x=350,y=90,width=250)
        
        
        contact=Label(register_frame,text="Contact No",font=("times new roman",13,"bold"),bg="white").place(x=30,y=120)
        self.txt_contact=Entry(register_frame,font=("times new roman",13),bg="#ECECEC")
        self.txt_contact.place(x=30,y=140,width=250)  
        
        
        email=Label(register_frame,text="Email",font=("times new roman",13,"bold"),bg="white").place(x=350,y=120)
        self.txt_email=Entry(register_frame,font=("times new roman",13),bg="#ECECEC")
        self.txt_email.place(x=350,y=140,width=250)
        
        question=Label(register_frame,text="Security Question",font=("times new roman",13,"bold"),bg="white").place(x=30,y=170)
        self.cmb_question=ttk.Combobox(register_frame,font=("times new roman",13),state='readonly',justify=CENTER)
        self.cmb_question['values']=("Select","Your favourite dish","Date of birth","Father Name","Birth place","Pet name")
        self.cmb_question.place(x=30,y=190,width=250)
        self.cmb_question.current(0)
        
        
        answer=Label(register_frame,text="Answer",font=("times new roman",13,"bold"),bg="white").place(x=350,y=170)
        self.txt_answer=Entry(register_frame,font=("times new roman",13),bg="#ECECEC")
        self.txt_answer.place(x=350,y=190,width=250)
        
        password=Label(register_frame,text="Password",font=("times new roman",13,"bold"),bg="white").place(x=30,y=220)
        self.txt_password=Entry(register_frame,font=("times new roman",13),bg="#ECECEC")
        self.txt_password.place(x=30,y=240,width=250)
        
        cpassword=Label(register_frame,text="Confrim Password",font=("times new roman",13,"bold"),bg="white").place(x=350,y=220)
        self.txt_cpassword=Entry(register_frame,font=("times new roman",13),bg="#ECECEC")
        self.txt_cpassword.place(x=350,y=240,width=250)       
        
        
        self.var_chk=IntVar()
        chk=Checkbutton(register_frame,text='I Agree the terms and Conditions',onvalue=1,offvalue=0,bg="white",font=("times new roman",13),variable=self.var_chk).place(x=30,y=290) 
        self.btn_image=ImageTk.PhotoImage(file="RegisterNow.jpg")
        btn_register=Button(register_frame,image=self.btn_image,cursor="hand2",command=self.register_data,bd=0).place(x=250,y=340)
        
        btn=Button(self.root,text="Sign In",command=self.login_window,font=("times new roman",18,"bold"),cursor="hand2",bd=0,bg="lightblue",activebackground="lightblue").place(x=135,y=480,width=180)

        
        
    def login_window(self):
        self.root.destroy()
        import Loginwindow
        
        
    def register_data(self):
        if self.txt_fname.get()=="" or self.txt_lname.get()=="" or self.txt_contact.get()=="" or self.txt_email.get()=="" or self.cmb_question.get()=="Select" or self.txt_answer.get()=="" or self.txt_password.get()=="" or self.txt_cpassword.get()=="":
            messagebox.showerror("Error","All Fields are required ",parent=self.root)
        elif self.txt_password.get()!=self.txt_cpassword.get():
            messagebox.showerror("Error","Password and Confrim Password should be same",parent=self.root)
        elif self.var_chk.get()==0:
            messagebox.showerror("Error","Please agree our terms and conditions",parent=self.root)
        else:
            try:
                con=mysql.connector.connect(host="localhost",user="root",passwd="Madhu12@",database='jobportal')
                cursor=con.cursor()
                cursor.execute("insert into register(f_name,l_name,contact,email,question,answer,password,cpassword) values(%s,%s,%s,%s,%s,%s,%s,%s)",
                           (self.txt_fname.get(),
                            self.txt_lname.get(),
                            self.txt_contact.get(),
                            self.txt_email.get(),
                            self.cmb_question.get(),
                            self.txt_answer.get(),
                            self.txt_password.get(),
                            self.txt_cpassword.get()                           
                           ))
                con.commit()
                con.close()
                messagebox.showinfo("Success","Register Successfull ",parent=self.root)
            except Exception as es:
                messagebox.showerror("Error",f"Error due to: {str(es)}",parent=self.root)
root=Tk()
obj=Register(root)
root.mainloop()
```
